```python
import requests
import json
import time
import os

# Set the base URL for your API
BASE_URL = "http://localhost:8000"  # Change this if your API is hosted elsewhere

def upload_video(file_path):
    url = f"{BASE_URL}/upload_video"
    files = {"file": open(file_path, "rb")}
    response = requests.post(url, files=files)
    return response.json()

def process_video(video_id):
    url = f"{BASE_URL}/process_video"
    data = {"video_id": video_id}
    response = requests.post(url, json=data)
    return response.json()

def generate_song(video_id, suno_prompt):
    url = f"{BASE_URL}/generate"
    data = {"video_id": video_id, "suno_prompt": suno_prompt}
    response = requests.post(url, json=data)
    return response.json()

def post_process_video(video_id):
    url = f"{BASE_URL}/post_process_video"
    data = {"video_id": video_id}
    response = requests.post(url, json=data)
    return response.json()

# Step 1: Upload Video
print("Step 1: Uploading video...")
video_path = "path/to/your/video.mp4"  # Replace with your video file path
upload_result = upload_video(video_path)
print(json.dumps(upload_result, indent=2))

video_id = upload_result["video_id"]

# Step 2: Process Video
print("\nStep 2: Processing video...")
process_result = process_video(video_id)
print(json.dumps(process_result, indent=2))

suno_prompt = process_result["suno_prompt"]

# Step 3: Generate Song
print("\nStep 3: Generating song...")
generate_result = generate_song(video_id, suno_prompt)
print(json.dumps(generate_result, indent=2))

song_path = generate_result["song_path"]

# Step 4: Post-process Video
print("\nStep 4: Post-processing video...")
post_process_result = post_process_video(video_id)
print(json.dumps(post_process_result, indent=2))

final_video_path = post_process_result["output_path"]

# Display results
print("\nResults:")
print(f"Original video: {video_path}")
print(f"Generated audio: {song_path}")
print(f"Final video: {final_video_path}")

# Play the generated audio
display(Audio(filename=song_path))

# Display the final video
display(Video(final_video_path))

print("\nProcess completed successfully!")
```
